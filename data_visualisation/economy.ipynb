```python
def get_money_one_row(row,numb_team):
    if numb_team == 1: 
        colmin = 9
        colmax = 39
    else:
        colmin = 40
        colmax = 70
    money_game = []
    for col in range(colmin,colmax):
        val = eco.iloc[row,col]
        if (val >= 0)==True:
            money_game.append(val)
        else:
            break 
    return money_game

def get_winner_rounds(row):
    winner_rounds = []
    for col in range(69,99):
        val = eco.iloc[row,col]
        if (val >= 0)==True:
            winner_rounds.append(val)
        else:
            break 
    return winner_rounds
```

```python
fig, axes = plt.subplots(1,2)

barWidth = 0.8
y = get_money_one_row(0,2)
r = range(len(y))

winner_per_round = get_winner_rounds(0)
rounds = np.arange(1,len(winner_per_round)+1)

axes[0].barh(r,y,height = barWidth, color = ['crimson' for i in y])
axes[1].plot(winner_per_round,rounds)
plt.grid = True
plt.show()
```
